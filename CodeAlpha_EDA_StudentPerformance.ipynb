```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
### reading an excel files 
df = pd.read_csv('C:\\Users\\TOSHIBA\\StudentsPerformance.csv')
print(df)

# Quick look at the data



```

```python
print(df.head())
print(df.info())
print(df.describe())
```

```python
# Gender distribution
sns.countplot(x='gender', data=df)
plt.title("Gender Distribution")
plt.show()


```

```python
# Average scores by gender
avg_scores = df.groupby('gender')[['math score', 'reading score', 'writing score']].mean()
print(avg_scores)

```

```python
# Boxplot of math scores by gender
sns.boxplot(x='gender', y='math score', data=df)
plt.title("Math Scores by Gender")
plt.show()

```

```python
# Test preparation vs scores
sns.boxplot(x='test preparation course', y='reading score', data=df)
plt.title("Reading Score vs Test Preparation")
plt.show()

```

```python
# Correlation heatmap
sns.heatmap(df[['math score', 'reading score', 'writing score']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Between Scores")
plt.show()
```
